```python
if __name__ == "__main__":
  print("start")

class Animal:
  def __init__(self,name,weight):
    self.name = name
    self.weight = weight
  
  def feed(self):
    print(f'{self.name} накормлен')

class MilkA(Animal):
  def milk(self):
    print(f'{self.name} подоено(а)')

class WoolA(Animal):
  def shear(self):
    print(f"{self.name} подстрижен")

class EggA(Animal):
  def collect(self):
    print(f"{self.name} собраны яйца")

class Goose(EggA):
  def __init__(self,name,weight):
    super().__init__(name,weight)
    self.voice = "щшшшшшшш"

class Cow(MilkA):
  def __init__(self,name,weight):
    super().__init__(name,weight)
    self.voice = "му-му"

class Sheep(WoolA):
  def __init__(self,name,weight):
    super().__init__(name,weight)
    self.voice = "бее"

class Chicken(EggA):
  def __init__(self,name,weight):
    super().__init__(name,weight)
    self.voice = "ко-ко-ко-ко"

class Duck(EggA):
  def __init__(self,name,weight):
    super().__init__(name,weight)
    self.voice = "кря-кря"

class Goat(MilkA):
  def __init__(self,name,weight):
    super().__init__(name,weight)
    self.voice = "меееее"

#Создаем экземпляры
goose1 = Goose("Серый",5)
goose2 = Goose("Белый",6)
cow = Cow("Манька",450)
sheep1 = Sheep("Барашек", 50)
sheep2 = Sheep("Кудрявый", 55)
chicken1 = Chicken("Ко-ко",2)
chicken2 = Chicken("Кукареку",3)
duck = Duck("Кряква",8)
goat1 = Goat("Рога",70)
goat2 = Goat("Копыта",75)

#Накормить
print("Кормим:")
goose1.feed()
goose2.feed()
cow.feed()
sheep1.feed()
sheep2.feed()
chicken1.feed()
chicken2.feed()
duck.feed()
goat1.feed()
goat2.feed()

#Взаимодействия
print("Взаимодействуем:")
cow.milk()
goat1.milk()
goat2.milk()
sheep1.shear()
sheep2.shear()
chicken1.collect()
chicken2.collect()
goose1.collect()
goose2.collect()
duck.collect()

#Общий вес
print("Считаем общий вес:")
animals =[goose1,goose2,sheep1,sheep2,chicken1,chicken2,cow,duck,goat1,goat2]
total = sum(animal.weight for animal in animals)
print(f"Общий вес животных: {total} килограмм")

#Самое тяжелое животное
print("Определяем самое тяжелое животное: ")
heaviest = max(animal.weight for animal in animals)
print(f"Самое тяжелое животное: {heaviest}")
for animal in animals:
  if heaviest == animal.weight:
    print(animal.name)

#различаем по голосам
for animal in animals:
  print(f"{animal.name}: {animal.voice}")
```
